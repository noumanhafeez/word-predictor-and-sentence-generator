```python
import re
def preprocess_text(text):
    """
    Preprocess text for N-gram language modeling.

    Steps:
    1. Remove Project Gutenberg header and footer
    2. Convert to lowercase
    3. Replace newline characters with spaces
    4. Split into sentences using ., ?, !
    5. Add <s> and </s> boundary tokens
    6. Remove punctuation
    7. Tokenize using whitespace
    """

    # -------------------------------------------------
    # 1. Remove Project Gutenberg header and footer
    # -------------------------------------------------

    # Remove header (everything before START marker)
    text = re.sub(
        r'.*?\*\*\*\s*start of the project gutenberg ebook.*?\*\*\*',
        '',
        text,
        flags=re.IGNORECASE | re.DOTALL
    )

    # Remove footer (everything after END marker)
    text = re.sub(
        r'\*\*\*\s*end of the project gutenberg ebook.*',
        '',
        text,
        flags=re.IGNORECASE | re.DOTALL
    )

    # -------------------------------------------------
    # 2. Convert to lowercase
    # -------------------------------------------------
    text = text.lower()

    # -------------------------------------------------
    # 3. Replace newline characters with spaces
    # -------------------------------------------------
    text = re.sub(r'[\r\n]+', ' ', text)

    # -------------------------------------------------
    # 4. Split text into sentences BEFORE removing punctuation
    #    using ., ?, !
    # -------------------------------------------------
    sentences = re.split(r'[.!?]+', text)

    token_list = []

    for sentence in sentences:
        sentence = sentence.strip()

        if not sentence:
            continue

        # -------------------------------------------------
        # 4a. Remove punctuation (after sentence splitting)
        # -------------------------------------------------
        sentence = re.sub(r'[^\w\s]', '', sentence)

        # -------------------------------------------------
        # 5. Tokenize by whitespace
        # -------------------------------------------------
        words = sentence.split()

        if not words:
            continue

        # -------------------------------------------------
        # 6. Add boundary tokens
        # -------------------------------------------------
        token_list.append("<s>")
        token_list.extend(words)
        token_list.append("</s>")



    return token_list
```

```python
import re
import string

def preprocess_for_ngrams(text):


    # 1. Remove Project Gutenberg header
    text = re.sub(r"The Project Gutenberg.*SAM'L GABRIEL SONS & COMPANY\s+NEW YORK", '', text, flags=re.I | re.S)

    # 2. Remove Project Gutenberg footer (everything after [Illustration] or END marker)
    text = re.sub(r"\*\*\* end of the project.*", '', text, flags=re.I | re.S)

    # 3. Lowercase
    text = text.lower()

    # 4. Replace newlines and tabs with space
    text = re.sub(r'\s+', ' ', text)

    # 5. Split into sentences using ., ?, !
    sentences = re.split(r'[.!?]+', text)
    #sentences = sentence_endings.split(text)


    # 6. Add boundary tokens and remove punctuation except apostrophes
    tokens = []

    for sentence in sentences:
        # Split sentence into words
        sentence_tokens = sentence.split()

        # Clean words and remove empty strings
        clean_tokens = []
        for w in sentence_tokens:
            w_clean = re.sub(r"[^\w']+", '', w)  # keep letters, digits, apostrophes
            if w_clean:
                clean_tokens.append(w_clean)

        # Add boundary tokens if sentence is not empty
        if clean_tokens:
            tokens += ['<s>'] + clean_tokens + ['</s>']
    return tokens
```

```python
from collections import Counter

# -----------------------------
# 1. Preprocessed words (ngram)
# -----------------------------
words = ngram  # your preprocessed word list

# -----------------------------
# 2. Build bigram and trigram counts
# -----------------------------
bigram_counts = Counter((words[i], words[i+1]) for i in range(len(words)-1))
trigram_counts = Counter((words[i], words[i+1], words[i+2]) for i in range(len(words)-2))

# -----------------------------
# 3. Bigram prediction
# -----------------------------
def predict_next_bigram(word):
    max_count = 0
    next_word = None
    for (w1, w2), count in bigram_counts.items():
        if w1 == word:
            if count > max_count:
                max_count = count
                next_word = w2
    return next_word

# -----------------------------
# 4. Trigram prediction with backoff to bigram
# -----------------------------
def predict_next_trigram(word1, word2):
    max_count = 0
    next_word = None
    for (w1, w2, w3), count in trigram_counts.items():
        if w1 == word1 and w2 == word2:
            if count > max_count:
                max_count = count
                next_word = w3

    # Backoff: if trigram fails, use bigram
    if next_word is None:
        return predict_next_bigram(word2)
    return next_word

# -----------------------------
# 5. Example usage
# -----------------------------
print("Bigram prediction after 'looked':", predict_next_bigram('very'))
print("Trigram prediction after 'very looked':", predict_next_trigram('very', 'looked'))
```

```python
# -----------------------------
# 1. Preprocessed words (ngram)
# -----------------------------
words = ngram  # your preprocessed word list

# -----------------------------
# 2. Build bigram and trigram lists
# -----------------------------
bigrams = [(words[i], words[i+1]) for i in range(len(words)-1)]
trigrams = [(words[i], words[i+1], words[i+2]) for i in range(len(words)-2)]

# -----------------------------
# 3. Bigram prediction
# -----------------------------
def predict_next_bigram(word):
    candidates = [w2 for w1, w2 in bigrams if w1 == word]
    if candidates:
        # most frequent = the one with maximum occurrences
        return max(set(candidates), key=candidates.count)
    return None

# -----------------------------
# 4. Unigram prediction (most common word)
# -----------------------------
def predict_next_unigram():
    return max(set(words), key=words.count)

# -----------------------------
# 5. Trigram prediction with backoff
# -----------------------------
def predict_next_trigram(word1, word2):
    candidates = [w3 for w1, w2, w3 in trigrams if w1 == word1 and w2 == word2]
    if candidates:
        return max(set(candidates), key=candidates.count)

    print(f"Trigram '{word1} {word2}' not found. Trying bigram with '{word1}'...")
    next_word = predict_next_bigram(word1)

    if next_word is None:
        print(f"Bigram '{word1}' not found. Using most common unigram as prediction...")
        next_word = predict_next_unigram()

    return next_word

# -----------------------------
# 6. User input loop
# -----------------------------
while True:
    choice = input("Enter '1' for bigram, '2' for trigram prediction, or 'q' to quit: ").strip()

    if choice == '1':
        word = input("Enter a word for bigram prediction: ").strip()
        next_word = predict_next_bigram(word)
        if next_word is None:
            print(f"Word '{word}' not found in bigram. Using most common unigram.")
            next_word = predict_next_unigram()
        print(f"Predicted next word: {next_word}\n")

    elif choice == '2':
        word1 = input("Enter the first word: ").strip()
        word2 = input("Enter the second word: ").strip()
        next_word = predict_next_trigram(word1, word2)
        print(f"Predicted next word: {next_word}\n")

    elif choice.lower() == 'q':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter 1, 2, or q.\n")
```

```python
import random

# -----------------------------
# 1. Generate next word with trigram backoff
# -----------------------------
def predict_next_trigram(word1, word2):
    # Try trigram first
    candidates = [w3 for w1, w2, w3 in trigrams if w1 == word1 and w2 == word2]
    if candidates:
        return max(set(candidates), key=candidates.count)

    # Backoff to bigram
    candidates_bigram = [w2 for w1, w2 in bigrams if w1 == word2]
    if candidates_bigram:
        return max(set(candidates_bigram), key=candidates_bigram.count)

    # Fallback to unigram
    return max(set(words), key=words.count)

# -----------------------------
# 2. Generate a sentence automatically
# -----------------------------
def generate_sentence(max_len=20):
    sentence = ['<s>']  # start-of-sentence token

    # Pick first word after <s> using bigram
    first_word_candidates = [w2 for w1, w2 in bigrams if w1 == '<s>']
    if not first_word_candidates:
        first_word = max(set(words), key=words.count)
    else:
        first_word = random.choice(first_word_candidates)

    sentence.append(first_word)

    # Pick second word using trigram (or bigram fallback)
    second_word_candidates = [w2 for w1, w2 in bigrams if w1 == sentence[-1]]
    if second_word_candidates:
        second_word = random.choice(second_word_candidates)
    else:
        second_word = max(set(words), key=words.count)

    sentence.append(second_word)

    # Generate remaining words
    while len(sentence) < max_len:
        next_word = predict_next_trigram(sentence[-2], sentence[-1])
        if next_word == '</s>':
            break
        sentence.append(next_word)

    # Remove start token and join sentence
    sentence = [w for w in sentence if w != '<s>']
    return ' '.join(sentence)

# -----------------------------
# 3. Interactive loop for sentence generation
# -----------------------------
print("Automatic Sentence Generator (Trigram + Backoff)!")
print("Type 'q' to quit.\n")

while True:
    cmd = input("Press Enter to generate a sentence or 'q' to quit: ").strip()
    if cmd.lower() == 'q':
        print("Exiting...")
        break
    generated = generate_sentence(max_len=30)
    print(f"Generated Sentence: {generated}\n")
```
